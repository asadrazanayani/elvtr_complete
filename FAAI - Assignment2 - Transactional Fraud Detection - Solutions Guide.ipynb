### 1. Introduction
- In this example, we will use the real-world 'creditcard.csv' dataset to walk through a basic data science workflow for fraud detection.

```python
# 1. Library Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# 2. Load the Dataset
df = pd.read_csv('creditcard.csv')

# Quick look at the data
print("Dataset shape:", df.shape)
print("\nDataset summary statistics:")
print(df.describe())

# Check class distribution
print("\nClass distribution:")
print(df['Class'].value_counts(normalize=True))

# 3. Prepare the Data
# Drop 'Time' feature, use 'Amount' and PCA features
X = df.drop(['Time', 'Class'], axis=1)
y = df['Class']

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 4. Build a Simple Decision Tree Model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# 5. Evaluate the Model
y_pred = model.predict(X_test)

print("\nModel Evaluation Metrics:")
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1-Score:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.title('Confusion Matrix')
plt.show()
```

# 🧪 Train/Test Split: Adjusting the Sampling Strategy

Now that we've performed extensive exploratory data analysis, it's time to prepare the data for model development.

In real-world fraud detection, careful splitting of training and testing sets is crucial because:

- The dataset is **heavily imbalanced**.
- Random splits may inadvertently distort the class distribution in small datasets.
- Maintaining proportional class balance helps us evaluate model generalization properly.

In this assignment:

- We previously used a **70/30 train/test split**.
- We will now update this to an **80/20 split** to allow slightly more training data for model development while still retaining sufficient holdout data for testing.

---

### Why adjust to 80/20?

- With very rare fraud cases, we benefit from maximizing training examples.
- We still retain enough fraud cases in the test set to evaluate model performance.
- This reflects a common practical approach used in production settings.

---

### Visualizing Before and After:

- We will create bar charts to visualize the number of fraud and legitimate cases in both training and testing sets before and after the split adjustment.
- This helps us verify that **class imbalance remains consistent across splits**, which is critical for reliable model validation.


## ⚠ Addressing Class Imbalance Before Modeling

Following our Exploratory Data Analysis (EDA), we observe a severe class imbalance in the dataset:

- Legitimate transactions represent over 99.8% of the data.
- Fraudulent transactions account for less than 0.2% of all observations.

Such extreme imbalance poses significant challenges for machine learning models, as they may become biased towards predicting the dominant class, failing to identify rare fraud cases.

### Train-Test Split Adjustment

For this assignment, we are modifying the default 70/30 split to an **80/20 stratified split**, preserving the class proportions during partitioning:

- **80%** of the data will be used for model training.
- **20%** will be held out as a test set for evaluation.

Stratification ensures that both training and test sets maintain a representative proportion of fraud cases.

### Introducing SMOTE (Synthetic Minority Oversampling Technique)

Rather than discarding legitimate transactions through random undersampling, we leverage **SMOTE** to synthetically generate new fraud samples:

- SMOTE interpolates between existing fraud cases to create realistic synthetic observations.
- This allows us to retain all legitimate training data while augmenting the minority fraud class.
- We specifically balance the training set to achieve an **80/20 class distribution** (80% legitimate, 20% fraud) — a controlled, more realistic scenario for training fraud detection models.

> **Important Note:**  
> SMOTE is applied **only to the training set** to avoid data leakage into the test set.


```python
# -----------------------------------------------
# FULL 80/20 Train-Test Split + SMOTE Balancing Code
# -----------------------------------------------

import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from collections import Counter
import plotly.graph_objects as go

# STEP 1: Start with original dataset (df) where 'Class' is your target
X = df.drop(columns=['Class'])
y = df['Class']

# STEP 2: Perform 80/20 stratified Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42
)

# STEP 3: Calculate target fraud count to achieve 80/20 class balance in training set
desired_fraud_ratio = 0.20  # 20% fraud, 80% legitimate
current_legit_count = y_train.value_counts()[0]
target_fraud_count = int(current_legit_count * desired_fraud_ratio / (1 - desired_fraud_ratio))

print(f"Target fraud count after SMOTE: {target_fraud_count}")

# STEP 4: Apply SMOTE with controlled sampling strategy
smote = SMOTE(sampling_strategy={1: target_fraud_count}, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# STEP 5: Verify class balance after SMOTE
print("Original Training Set Class Distribution:")
print(y_train.value_counts())
print("\nAfter SMOTE Balancing:")
print(Counter(y_train_smote))

# STEP 6: Visualize class balance after SMOTE
counts = pd.Series(y_train_smote).value_counts().sort_index()

fig = go.Figure(data=[
    go.Bar(x=['Legitimate', 'Fraud'], y=[counts[0], counts[1]], marker_color='purple')
])

fig.update_layout(
    title="Training Set Class Distribution After SMOTE Balancing to 80/20 Ratio",
    yaxis_title="Count of Transactions",
    width=600, height=400
)

fig.show()
```

## ✅ Post-SMOTE Balancing Results

The output confirms that SMOTE has successfully generated synthetic fraud samples to achieve the targeted 80/20 class balance within the training set:

- **Legitimate transactions:** 227,451  
- **Fraudulent transactions (after SMOTE):** 56,862  
- This results in approximately 20% fraud representation in the training data.

The bar chart provides a clear visual confirmation of the updated class distribution.

By applying SMOTE in this controlled manner, we preserve the full signal of legitimate transactions while providing the model with sufficient fraud examples to learn complex decision boundaries. This approach strikes a balance between maintaining data integrity and addressing class imbalance — a key challenge in real-world fraud detection modeling.


## 📊 Model Comparison with Multiple Algorithms

With the dataset fully preprocessed, balanced, and scaled, we now proceed to develop and compare several machine learning models commonly applied in fraud detection.

### Models included in this comparison:

- **Decision Tree Classifier**  
- **Logistic Regression** (requires feature scaling)  
- **Random Forest Classifier**  
- **XGBoost Classifier**  
- **Neural Network (Multi-Layer Perceptron)** (requires feature scaling)

Each model will be tuned using basic hyperparameter grid search, and evaluated against the untouched test set to ensure honest performance assessment.

---

### ⚠ Sampling for Practical Model Development

Our SMOTE-balanced training dataset contains approximately 283,000 rows after oversampling. Performing full grid search hyperparameter tuning with multiple models and cross-validation folds on this full dataset would require extensive computational time.

To ensure timely and efficient model training while still maintaining pedagogical value, we randomly sample **20,000 balanced observations** from the SMOTE training set for model development.

> This controlled sampling allows us to:
> - Preserve class balance between legitimate and fraud transactions.
> - Drastically reduce model training time.
> - Allow full model comparison across multiple algorithms in a classroom or resource-limited environment.

---

### ✅ Evaluation Metrics:

For each model, we will report:
- Area Under the ROC Curve (**AUC**)
- **Precision**
- **Recall**
- **F1 Score**

These metrics will allow us to compare how well each algorithm performs in detecting fraud while balancing false positives and false negatives.

---


```python
# -----------------------------------------
# Fast Model Comparison with Sampling & GridSearch
# -----------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample

# ------------------------------------------------------------
# Assume you already have:
# - X_train_smote, y_train_smote (SMOTE balanced train set)
# - X_test, y_test (original untouched test set)
# ------------------------------------------------------------

# STEP 1: Sample the SMOTE-balanced training set for faster modeling
X_train_sampled, y_train_sampled = resample(
    X_train_smote, y_train_smote, 
    n_samples=20000, random_state=42
)

# STEP 2: Scale features where appropriate
scaler = StandardScaler()
X_train_sampled_scaled = scaler.fit_transform(X_train_sampled)
X_test_scaled = scaler.transform(X_test)

# Create dictionary to store model results
results = {
    'Model': [],
    'AUC': [],
    'Precision': [],
    'Recall': [],
    'F1': []
}

# ---------------------------
# 1. Decision Tree (unscaled)
# ---------------------------
dt_param_grid = {'max_depth': [5, 10]}
dt_grid = GridSearchCV(DecisionTreeClassifier(random_state=42), dt_param_grid, cv=3, scoring='roc_auc')
dt_grid.fit(X_train_sampled, y_train_sampled)
dt_best = dt_grid.best_estimator_
y_pred_dt = dt_best.predict(X_test)

results['Model'].append('Decision Tree')
results['AUC'].append(roc_auc_score(y_test, dt_best.predict_proba(X_test)[:,1]))
results['Precision'].append(precision_score(y_test, y_pred_dt))
results['Recall'].append(recall_score(y_test, y_pred_dt))
results['F1'].append(f1_score(y_test, y_pred_dt))

# ---------------------------
# 2. Logistic Regression (scaled)
# ---------------------------
lr_param_grid = {'C': [0.1, 1]}
lr_grid = GridSearchCV(LogisticRegression(max_iter=2000, random_state=42), lr_param_grid, cv=3, scoring='roc_auc')
lr_grid.fit(X_train_sampled_scaled, y_train_sampled)
lr_best = lr_grid.best_estimator_
y_pred_lr = lr_best.predict(X_test_scaled)

results['Model'].append('Logistic Regression')
results['AUC'].append(roc_auc_score(y_test, lr_best.predict_proba(X_test_scaled)[:,1]))
results['Precision'].append(precision_score(y_test, y_pred_lr))
results['Recall'].append(recall_score(y_test, y_pred_lr))
results['F1'].append(f1_score(y_test, y_pred_lr))

# ---------------------------
# 3. Random Forest (unscaled)
# ---------------------------
rf_param_grid = {'n_estimators': [100], 'max_depth': [10]}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_param_grid, cv=3, scoring='roc_auc')
rf_grid.fit(X_train_sampled, y_train_sampled)
rf_best = rf_grid.best_estimator_
y_pred_rf = rf_best.predict(X_test)

results['Model'].append('Random Forest')
results['AUC'].append(roc_auc_score(y_test, rf_best.predict_proba(X_test)[:,1]))
results['Precision'].append(precision_score(y_test, y_pred_rf))
results['Recall'].append(recall_score(y_test, y_pred_rf))
results['F1'].append(f1_score(y_test, y_pred_rf))

# ---------------------------
# 4. XGBoost (unscaled)
# ---------------------------
xgb_param_grid = {'n_estimators': [100], 'max_depth': [5]}
xgb_grid = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), 
                        xgb_param_grid, cv=3, scoring='roc_auc')
xgb_grid.fit(X_train_sampled, y_train_sampled)
xgb_best = xgb_grid.best_estimator_
y_pred_xgb = xgb_best.predict(X_test)

results['Model'].append('XGBoost')
results['AUC'].append(roc_auc_score(y_test, xgb_best.predict_proba(X_test)[:,1]))
results['Precision'].append(precision_score(y_test, y_pred_xgb))
results['Recall'].append(recall_score(y_test, y_pred_xgb))
results['F1'].append(f1_score(y_test, y_pred_xgb))

# ---------------------------
# 5. Neural Network (scaled)
# ---------------------------
mlp_param_grid = {'hidden_layer_sizes': [(50,)], 'alpha':[0.0001]}
mlp_grid = GridSearchCV(MLPClassifier(max_iter=1000, random_state=42), mlp_param_grid, cv=3, scoring='roc_auc')
mlp_grid.fit(X_train_sampled_scaled, y_train_sampled)
mlp_best = mlp_grid.best_estimator_
y_pred_mlp = mlp_best.predict(X_test_scaled)

results['Model'].append('Neural Network')
results['AUC'].append(roc_auc_score(y_test, mlp_best.predict_proba(X_test_scaled)[:,1]))
results['Precision'].append(precision_score(y_test, y_pred_mlp))
results['Recall'].append(recall_score(y_test, y_pred_mlp))
results['F1'].append(f1_score(y_test, y_pred_mlp))

# ---------------------------
# Create Results DataFrame
# ---------------------------
results_df = pd.DataFrame(results)
display(results_df)

# ---------------------------
# Visualize Performance Metrics
# ---------------------------
fig, axs = plt.subplots(2, 2, figsize=(12,10))
sns.barplot(x='Model', y='AUC', data=results_df, ax=axs[0,0]).set_title("AUC")
sns.barplot(x='Model', y='Precision', data=results_df, ax=axs[0,1]).set_title("Precision")
sns.barplot(x='Model', y='Recall', data=results_df, ax=axs[1,0]).set_title("Recall")
sns.barplot(x='Model', y='F1', data=results_df, ax=axs[1,1]).set_title("F1 Score")
plt.tight_layout()
plt.show()
```

However, AUC is only part of the story. In fraud detection, **Precision** and **Recall** have much stronger business implications.

---

### ⚠ Precision vs. Recall Tradeoff

- **Recall** answers:  
  > "Of all the actual fraud cases, how many did we correctly identify?"

- **Precision** answers:  
  > "Of all the transactions flagged as fraud, how many were actually fraud?"

---

### Why False Positives Matter

Because this is a **fraud detection system**, the models will trigger actions such as:
- Flagging transactions for manual review
- Declining transactions
- Freezing accounts
- Inconveniencing legitimate customers

> **False Positives = Legitimate customers flagged incorrectly as fraud.**  
> High False Positive rates create operational costs, customer frustration, and reputational risk.

## 🔬 Partial Dependence Plots (PDP) with Volume

While feature importance helps us rank which variables are most predictive, it does not explain *how* each feature influences the model’s predictions.

Partial Dependence Plots (PDP) allow us to visualize the marginal effect of a single feature on the model's predicted fraud probability, holding all other features constant. 

However, **standard PDP plots can sometimes be misleading** when data density is highly imbalanced across feature ranges. Therefore, we will combine PDP plots with an additional volume overlay that shows the actual frequency (or count) of data points at each value of the feature. This gives us better intuition about:

- The shape of the relationship between feature value and predicted fraud probability.
- Where most of the real data points are concentrated.
- Regions where the model may be extrapolating based on sparse data.

We will run PDP plots for the top contributing features identified by our Random Forest model.


```python
# ---------------------------------------------------------------
# Full PDP Visualization (Top 15 Features) with Volume Overlay
# ---------------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.inspection import partial_dependence
import numpy as np

# Select top N features (we'll do top 15)
top_n = 15
top_features = feature_importances_df['Feature'].values[:top_n]

# Determine grid layout dynamically
cols = 3
rows = int(np.ceil(top_n / cols))

fig = plt.figure(figsize=(cols*6, rows*5))
gs = gridspec.GridSpec(rows, cols)

for idx, feature in enumerate(top_features):
    row = idx // cols
    col = idx % cols
    ax = fig.add_subplot(gs[row, col])
    
    # Calculate PDP full range
    pd_result = partial_dependence(rf_best, X_test, [feature], kind='average')
    values = pd_result['values'][0]
    averaged_preds = pd_result['average'][0]
    
    # Apply 5th-95th percentile filtering
    lower_bound = np.percentile(X_test[feature], 5)
    upper_bound = np.percentile(X_test[feature], 95)
    mask = (values >= lower_bound) & (values <= upper_bound)
    values_filtered = values[mask]
    preds_filtered = averaged_preds[mask]
    
    # PDP Line
    ax.plot(values_filtered, preds_filtered, color='purple', lw=2)
    ax.set_xlabel(feature)
    ax.set_ylabel("Predicted Fraud Probability")
    ax.set_title(f"PDP: {feature}")
    
    # Overlay histogram
    ax2 = ax.twinx()
    ax2.hist(X_test[feature], bins=30, color='lightgray', alpha=0.5, range=(lower_bound, upper_bound))
    ax2.set_ylabel("Transaction Volume", color='gray')
    ax2.tick_params(axis='y', labelcolor='gray')

plt.suptitle("Partial Dependence Plots (Top 15 Features, Filtered 5th-95th Percentile)", fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
```

## Decile Analysis, Lift Table, and Model Performance Visualization

After evaluating model accuracy using traditional metrics like AUC, Precision, Recall, and F1 score, it's critical to assess how well the model ranks transactions based on predicted fraud probability. This is particularly valuable for business teams who want to operationalize fraud detection with limited resources — for example, only investigating the top X% highest risk transactions.

We will create:

- A **Lift Table** broken into 10 deciles (10% buckets of the population), sorted by predicted fraud risk.
- For each decile, we will compute:
    - Total observations
    - Actual fraud count
    - Expected fraud count (based on model probabilities)
    - Actual and expected fraud rates
    - Cumulative fraud counts
    - Cumulative non-fraud counts
    - Kolmogorov-Smirnov (KS) statistic
    - Lift score compared to overall base fraud rate.

- We will also generate:
    - A **Lift Chart** (model lift vs random selection)
    - A **Gain Chart** (cumulative capture of fraud across deciles)

This type of analysis provides powerful insight into how well the model separates high-risk vs low-risk transactions and helps quantify business value.


```python
# ---------------------------------------------
# CALIBRATION + FIXED BINNING FOR LIFT TABLE
# ---------------------------------------------

from sklearn.calibration import CalibratedClassifierCV

# 1️⃣ Apply calibration using imbalanced test set
calibrator = CalibratedClassifierCV(estimator=rf_best, method='isotonic', cv='prefit')
calibrator.fit(X_test_scaled, y_test)

# 2️⃣ Generate calibrated probabilities
y_test_probs_calibrated = calibrator.predict_proba(X_test_scaled)[:, 1]

# 3️⃣ Build Lift Table with robust binning
test_data = pd.DataFrame({
    'y_true': y_test,
    'y_pred_proba': y_test_probs_calibrated
})

# Fix binning: rank to force 10 equal-sized bins
test_data['rank'] = test_data['y_pred_proba'].rank(method='first')
test_data['decile'] = pd.qcut(test_data['rank'], 10, labels=False) + 1

# Group by decile
lift_table = test_data.groupby('decile').agg(
    total_obs=('y_true', 'count'),
    actual_fraud=('y_true', 'sum'),
    expected_fraud=('y_pred_proba', 'sum')
).reset_index()

# Compute metrics
lift_table['actual_rate'] = lift_table['actual_fraud'] / lift_table['total_obs']
lift_table['expected_rate'] = lift_table['expected_fraud'] / lift_table['total_obs']
lift_table['cumulative_actual_fraud'] = lift_table['actual_fraud'].cumsum()
lift_table['cumulative_non_fraud'] = lift_table['total_obs'].cumsum() - lift_table['cumulative_actual_fraud']
lift_table['lift'] = lift_table['actual_rate'] / (test_data['y_true'].sum() / test_data['y_true'].count())
lift_table['cum_actual_pct'] = lift_table['cumulative_actual_fraud'] / test_data['y_true'].sum()
lift_table['cum_nonfraud_pct'] = lift_table['cumulative_non_fraud'] / (test_data['y_true'].count() - test_data['y_true'].sum())
lift_table['ks'] = abs(lift_table['cum_actual_pct'] - lift_table['cum_nonfraud_pct'])

# Display
pd.set_option('display.float_format', '{:.6f}'.format)
display(lift_table)
```

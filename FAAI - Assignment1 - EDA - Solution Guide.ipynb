## 📊 Histograms of Feature Distributions

### Why Use Histograms?

- **Histograms** allow us to see the **overall shape of the distribution** for each feature.
- This helps us identify:
  - Skewness
  - Multi-modality
  - Concentration zones
  - Potential need for feature transformation (e.g., normalization, log-scaling)
- Histograms complement boxplots by providing a more continuous view of how values are distributed across the entire population.

---

### Strategy for This Analysis

- We'll examine:
  - The distribution of `Amount` (both raw and log-transformed).
  - The distribution of several PCA features.
  - Compare distributions for legitimate vs fraudulent transactions side-by-side.

---


````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import skew, kurtosis

# Create list of all features
pca_features = [col for col in df.columns if col.startswith('V')]
all_features = pca_features + ['Amount']

# Set up subplots: 29 features → 15 rows x 2 columns
fig, axes = plt.subplots(nrows=15, ncols=2, figsize=(16, 60))
axes = axes.flatten()

for idx, feature in enumerate(all_features):
    
    # Select data and apply log transformation to Amount
    if feature == 'Amount':
        data = np.log1p(df[feature])
    else:
        data = df[feature]
    
    # Calculate statistics
    feature_skew = skew(data)
    feature_kurt = kurtosis(data)
    feature_std = np.std(data)
    
    # Plot histogram
    sns.histplot(data, bins=100, kde=True, ax=axes[idx], color='skyblue')
    axes[idx].set_title(f'{feature}', fontsize=14)
    axes[idx].set_xlabel('Value')
    axes[idx].set_ylabel('Frequency')
    
    # Annotate stats on plot
    stats_text = f'Skew: {feature_skew:.2f}\nKurtosis: {feature_kurt:.2f}\nStd: {feature_std:.2f}'
    axes[idx].text(0.95, 0.95, stats_text,
                   transform=axes[idx].transAxes,
                   verticalalignment='top',
                   horizontalalignment='right',
                   fontsize=10,
                   bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    
# Remove any empty subplots
for j in range(len(all_features), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
````

## 🔎 Analysis of Full Feature Histograms

### Skewness:

- Many PCA features exhibit strong skewness:
  - Several components are highly **left-skewed** (negative skew), with rare but extreme negative values.
  - Some features show moderate **right-skew**.
  - Skewed distributions suggest the presence of extreme but infrequent transaction behaviors, often useful in detecting fraud-related anomalies.

### Kurtosis:

- Several features display extremely high kurtosis (peakedness), indicating:
  - Sharp central peaks where most transactions cluster very tightly.
  - Heavy tails where extreme outliers occur.
- Examples:
  - `V7` shows kurtosis over 400.
  - `V5` and `V8` also exceed kurtosis values of 200.
- High kurtosis suggests that **most transactions look very "normal"**, but rare edge cases exist that could correlate with fraudulent activity.

### Standard Deviation:

- The standard deviations are generally small across PCA components.
- This tight concentration means that relatively small deviations from the mean may still represent meaningful departures from "normal" behavior.
- Features with low standard deviation but heavy tails are often highly informative for fraud models.

### Visual Shape Observations:

- Many features display strongly **non-normal distributions** even after PCA transformation.
- Certain components such as `V4` suggest possible multi-modal patterns, which may reflect the existence of multiple distinct behavioral clusters within the legitimate transaction population.
- The presence of heavy tails across many components supports the hypothesis that fraudsters may exploit specific rare behavioral regimes.

### Implications for Modeling:

- Skewness and kurtosis should inform **feature engineering**:
  - Consider transformations (e.g., log-scaling, winsorizing) for extreme variables.
  - Use robust scaling techniques that handle heavy-tailed data.
- These features may offer strong discriminative power for both supervised fraud classification and unsupervised anomaly detection algorithms.

## 📊 Correlation Heatmap

### Why Perform Correlation Analysis?

- **Correlation matrices** allow us to examine how strongly features are related to each other.
- Highly correlated features may:
  - Contain redundant information.
  - Inflate model variance (especially for linear models).
  - Introduce multicollinearity, which can destabilize certain models.
- In fraud detection, correlation analysis helps identify:
  - Redundant features that may be candidates for dimensionality reduction or elimination.
  - Feature groups that may be capturing related behavioral patterns.
  - Possible feature interactions worth engineering.

---

### How Does Correlation Analysis Help with Modeling?

- Many machine learning algorithms assume that features are independent (e.g., logistic regression, Naive Bayes). Highly correlated inputs violate this assumption.
- For **tree-based models (e.g., Random Forest, XGBoost)**, correlation is less problematic but still useful to understand:
  - Correlated features may "split" predictive power across variables.
  - Features with high correlation may compete for importance in tree-based splits.
- For **regularization methods (e.g., Lasso, Ridge)**, understanding correlation is critical since these models penalize coefficient sizes differently depending on correlation structure.
- In **unsupervised fraud detection (e.g., anomaly detection, clustering)**, highly correlated features may distort distance metrics or cluster definitions.

---

### Special Note for PCA Features:

- PCA transformation is designed to create orthogonal (uncorrelated) components.
- However, due to how PCA was originally performed on the raw data, or possible rescaling, **residual correlations may still exist**.
- Analyzing correlations among PCA features helps confirm whether the transformation was fully effective, or whether subtle interdependencies remain.

---

### What We Expect to See:

- Ideally, most PCA features should exhibit low pairwise correlations.
- Any strong correlations that remain may point to:
  - Possible dimensionality reduction opportunities.
  - Underlying behavioral linkages between transaction features.


````python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix for all features (excluding Class)
feature_cols = [col for col in df.columns if col != 'Class']
corr = df[feature_cols].corr()

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Draw the heatmap with annotations
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt=".2f", linewidths=0.5)

plt.title('Correlation Heatmap of Features', fontsize=18)
plt.show()
````

## 📈 Extra Credit Task 1: Plot Fraud Rate Over Time

Fraud events often occur in bursts rather than being evenly distributed over time. By visualizing fraud rate across time windows, we may detect:

- Sudden bursts of fraudulent activity.
- Stable periods with low or zero fraud.
- Potential patterns that could indicate coordinated fraud attacks or system vulnerabilities.

In this task, we will:

- Aggregate the data into equal time bins.
- Compute fraud rate within each bin.
- Visualize the fraud rate as a function of time.

This type of time-series fraud analysis can be highly valuable for real-time monitoring and early warning detection systems.


````python
import pandas as pd
import plotly.express as px

# Assuming your dataset is already loaded as 'df'
# and the Time column is in seconds since the first transaction.

# First, define the size of time bins (in seconds)
# For example, let's bin every 6 hours (21600 seconds)
bin_size = 21600  

# Create a new column for time bins
df['Time_Bin'] = (df['Time'] // bin_size) * bin_size

# Group by time bin and compute fraud rate
fraud_rate_over_time = df.groupby('Time_Bin')['Class'].agg(['count', 'sum'])
fraud_rate_over_time['Fraud_Rate'] = (fraud_rate_over_time['sum'] / fraud_rate_over_time['count']) * 100

# Reset index for plotting
fraud_rate_over_time.reset_index(inplace=True)

# Create the plot
fig = px.line(
    fraud_rate_over_time,
    x='Time_Bin',
    y='Fraud_Rate',
    title='Fraud Rate Over Time',
    labels={'Time_Bin': 'Time (binned)', 'Fraud_Rate': 'Fraud Rate (%)'},
    markers=True
)

fig.update_layout(width=900, height=500)
fig.show()
````

### 📊 Analysis: Fraud Rate Over Time

The plot of fraud rate over time reveals several important insights:

- The fraud rate is **highly variable across different time windows**. We observe noticeable spikes and valleys rather than a flat, stable rate.
- There is a significant **spike around the 80,000-second mark**, where the fraud rate exceeds 0.6%. This indicates a potential fraud burst or cluster of coordinated fraudulent activity.
- After this peak, the fraud rate drops sharply and remains relatively lower, but still fluctuates, suggesting periods of both high and low fraud activity.
- These bursts of fraud may correspond to:
  - Testing windows where fraudsters probe the system.
  - Coordinated attacks exploiting system weaknesses during specific time periods.
  - Temporary breakdowns in fraud prevention mechanisms.

**Why this matters:**

- Fraud is often **non-stationary** — it changes over time.
- Static models trained on historical averages may **miss emerging fraud patterns**.
- Incorporating **time-dependent features** or building models capable of detecting fraud bursts can improve detection accuracy.
- This type of temporal fraud analysis can serve as an important **real-time monitoring tool** for transaction systems.

**Next steps:**

- Consider zooming into these fraud burst windows to analyze feature values during spikes.
- Investigate potential connections between fraud clusters and transaction types, amounts, or other contextual variables.

````python
import pandas as pd
import plotly.express as px

# Define amount bins (these can be adjusted based on your data distribution)
amount_bins = [0, 10, 50, 100, 500, 1000, 5000, 10000, df['Amount'].max()]
bin_labels = ['0-10', '10-50', '50-100', '100-500', '500-1000', '1000-5000', '5000-10000', '10000+']

# Create binned amount column
df['Amount_Bin'] = pd.cut(df['Amount'], bins=amount_bins, labels=bin_labels, include_lowest=True)

# Group by amount bins and compute fraud rate
fraud_by_amount = df.groupby('Amount_Bin')['Class'].agg(['count', 'sum'])
fraud_by_amount['Fraud_Rate'] = (fraud_by_amount['sum'] / fraud_by_amount['count']) * 100
fraud_by_amount.reset_index(inplace=True)

# Plot
fig = px.bar(
    fraud_by_amount, 
    x='Amount_Bin', 
    y='Fraud_Rate', 
    title='Fraud Rate by Transaction Amount Bins',
    labels={'Fraud_Rate': 'Fraud Rate (%)', 'Amount_Bin': 'Transaction Amount Range'},
    text_auto=True
)

fig.update_layout(width=900, height=500)
fig.show()
````

- Color-code the scatterplots by fraud class to highlight class separation patterns.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Redefine feature list (PCA components only)
feature_list = [col for col in df.columns if col.startswith('V')]

# Subsample for pairplot performance
sample_size = 5000
df_sample = df.sample(n=sample_size, random_state=42)

# Calculate correlation matrix on sample
corr_matrix = df_sample[feature_list].corr().abs()

# Find most highly correlated pairs
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_pairs = (
    upper_triangle.stack()
    .sort_values(ascending=False)
    .reset_index()
)
high_corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']

# Select top correlated features
top_pairs = high_corr_pairs.head(3)
selected_features = pd.unique(top_pairs[['Feature_1', 'Feature_2']].values.ravel())

# Prepare pairplot data
pairplot_df = df_sample[list(selected_features) + ['Class']].copy()
pairplot_df['Class'] = pairplot_df['Class'].map({0: 'Legitimate', 1: 'Fraud'})

# Create pairplot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))
pairplot = sns.pairplot(pairplot_df, hue='Class', palette={'Legitimate': 'blue', 'Fraud': 'red'}, plot_kws={'alpha': 0.5})
pairplot.fig.suptitle("Seaborn Pairplot of Top Correlated PCA Features", y=1.02)
plt.show()
````

## 🧾 Extra Credit Task 5: Outlier Detection using Z-Score and IQR Methods

Fraudulent transactions often appear as statistical outliers relative to legitimate transactions. Simple outlier detection methods can offer valuable insight before advanced modeling:

- **Z-Score Method:** Flags observations that lie several standard deviations away from the mean.
- **IQR Method (Interquartile Range):** Flags observations that fall outside the typical interquartile spread (above Q3 + 1.5*IQR or below Q1 - 1.5*IQR).

In this task, we will:

- Apply both Z-Score and IQR outlier detection methods to the `Amount` variable.
- Compare how many legitimate vs. fraudulent transactions are flagged as outliers.
- Explore whether these simple statistical rules can capture some fraud cases.

Although these approaches are not sufficient for full fraud detection, they are **useful as preprocessing steps, exploratory diagnostics, or features in more complex fraud models**.


````python
import numpy as np
import pandas as pd

# Apply log transform to Amount for better scale stability
df['Log_Amount'] = np.log1p(df['Amount'])

### Z-SCORE METHOD ###

# Calculate Z-score
df['Z_Score'] = (df['Log_Amount'] - df['Log_Amount'].mean()) / df['Log_Amount'].std()

# Flag outliers based on Z-score threshold
z_threshold = 3  # Typical rule: beyond 3 standard deviations
df['Z_Outlier'] = np.where(abs(df['Z_Score']) > z_threshold, 1, 0)

# Summarize Z-score outliers by class
z_summary = df.groupby('Class')['Z_Outlier'].sum()
print("Z-Score Outliers Detected:")
print(z_summary)

### IQR METHOD ###

# Calculate IQR bounds
Q1 = df['Log_Amount'].quantile(0.25)
Q3 = df['Log_Amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag IQR outliers
df['IQR_Outlier'] = np.where((df['Log_Amount'] < lower_bound) | (df['Log_Amount'] > upper_bound), 1, 0)

# Summarize IQR outliers by class
iqr_summary = df.groupby('Class')['IQR_Outlier'].sum()
print("\nIQR Outliers Detected:")
print(iqr_summary)
````

### 🧾 Analysis: Outlier Detection using Z-Score and IQR Methods

#### Z-Score Results:

- **218 legitimate transactions** were flagged as Z-score outliers (beyond 3 standard deviations).
- **No fraudulent transactions** were captured using this rule.

#### IQR Results:

- **245 legitimate transactions** were flagged as IQR outliers.
- **Again, no fraudulent transactions** were flagged.

---

#### Interpretation:

- These simple statistical outlier detection methods failed to directly identify fraudulent transactions.
- This outcome illustrates a common challenge in fraud analytics:  
  - **Fraudulent transactions may not always appear as statistical outliers when using generic global thresholds.**
  - Fraud can occur inside the dense core of legitimate data (so-called *inliers*) rather than always being extreme.
  - Fraud often requires **contextual outlier detection** — where certain feature combinations are rare even if individual features are not extreme.
- Legitimate transactions that were flagged as outliers may still represent unusual customer behavior (large legitimate purchases, rare events), but are not fraud.

---

#### Key Takeaways:

- **Univariate outlier detection is insufficient for fraud detection.**
- More advanced multivariate approaches (e.g. clustering, Isolation Forest, LOF, autoencoders) are better suited for uncovering complex fraud patterns.
- However, these simple methods still serve as:
  - Useful diagnostics during EDA.
  - Early filtering or feature engineering steps in supervised pipelines.

---

#### Next-Level Insight:

- In real-world fraud detection systems, combining **statistical anomaly detection with supervised models** often delivers better results.
- Domain knowledge and engineered features (e.g., time windows, device fingerprints, merchant categories) are critical for building effective fraud scoring models.

````python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# --- Load your dataset ---
df = pd.read_csv('creditcard.csv')

# --- Page Title ---
st.title("🔥 Mind-Blowing Fraud Analytics Interactive Lab")

# --- Sidebar Controls ---
st.sidebar.title("🔧 Dashboard Controls")
feature_list = [col for col in df.columns if col.startswith('V')] + ['Amount']
selected_feature = st.sidebar.selectbox("Select Feature", feature_list)
log_transform = st.sidebar.checkbox("Log Transform (for Amount only)", value=False)
time_slider = st.sidebar.slider("Filter by Time Window", 
                                 int(df['Time'].min()), 
                                 int(df['Time'].max()), 
                                 (int(df['Time'].min()), int(df['Time'].max())))

# --- Apply Time Filter ---
df_filtered = df[(df['Time'] >= time_slider[0]) & (df['Time'] <= time_slider[1])]

# --- Tabs Layout ---
tab1, tab2, tab3, tab4, tab5, tab6, tab7 = st.tabs([
    "📊 Summary", "⚖️ Class Imbalance", "📦 Distribution", 
    "🌐 Correlation", "🧠 Anomaly Scoring", "🧪 3D PCA Scatter", "📂 Dataset Explorer"
])

# ========================
# TAB 1 — Summary Stats
# ========================
with tab1:
    st.header("📊 Summary Statistics")
    st.dataframe(df_filtered.describe().T)
    st.markdown("""
    **🔎 Insights:**
    - Fraud rate is extremely low, typical for real-world financial fraud.
    - Many PCA components show high skew and kurtosis, indicating sharp central clusters with rare outliers.
    - Careful feature engineering will be critical for modeling.
    """)

# ========================
# TAB 2 — Class Imbalance
# ========================
with tab2:
    st.header("⚖️ Class Imbalance")

    class_counts = df_filtered['Class'].value_counts().rename({0: 'Legitimate', 1: 'Fraud'})
    total = class_counts.sum()
    fraud_rate = round(class_counts.get('Fraud', 0) / total * 100, 4)

    bar_chart = go.Figure(
        data=[go.Bar(x=['Legitimate', 'Fraud'], y=class_counts.values, text=class_counts.values, textposition='auto')],
        layout_title_text=f"Class Distribution (Fraud Rate: {fraud_rate}%)"
    )
    st.plotly_chart(bar_chart)
    
    st.markdown("""
    **🔎 Insights:**
    - The extreme imbalance is typical in fraud detection datasets.
    - Any modeling strategy must carefully address this imbalance.
    - Sampling, cost-sensitive learning, or anomaly detection may be required.
    """)

# ========================
# TAB 3 — Distribution Analysis
# ========================
with tab3:
    st.header(f"📦 Distribution for {selected_feature}")

    if selected_feature == 'Amount' and log_transform:
        hist_data = np.log1p(df_filtered[selected_feature])
        xlabel = 'Log(Amount + 1)'
    else:
        hist_data = df_filtered[selected_feature]
        xlabel = selected_feature

    hist_fig = px.histogram(hist_data, nbins=100, title=f'Histogram of {xlabel}')
    st.plotly_chart(hist_fig)

    box_fig = px.box(df_filtered, x='Class', y=selected_feature, 
                     color=df_filtered['Class'].map({0: 'Legitimate', 1: 'Fraud'}),
                     points='all')
    box_fig.update_layout(
        xaxis=dict(tickmode='array', tickvals=[0, 1], ticktext=['Legitimate', 'Fraud']),
        title=f"Boxplot of {selected_feature} by Class"
    )
    st.plotly_chart(box_fig)

    st.markdown("""
    **🔎 Insights:**
    - Many features show distinct value ranges for fraudulent vs legitimate transactions.
    - Outliers are prevalent in legitimate transactions, while fraud often shows tighter clustering.
    """)

# ========================
# TAB 4 — Correlation Heatmap
# ========================
with tab4:
    st.header("🌐 Correlation Heatmap")

    corr = df_filtered[feature_list].corr()
    fig_heatmap, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr, cmap='coolwarm', ax=ax)
    st.pyplot(fig_heatmap)

    st.markdown("""
    **🔎 Insights:**
    - PCA features are largely uncorrelated as expected.
    - Minimal multicollinearity is present, ideal for many ML models.
    """)

# ========================
# TAB 5 — Anomaly Scoring
# ========================
with tab5:
    st.header("🧠 Anomaly Detection with Isolation Forest")

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(df_filtered[feature_list + ['Amount']])

    iso_forest = IsolationForest(contamination=0.005, random_state=42)
    df_filtered['Anomaly_Score'] = iso_forest.fit_predict(scaled_features)
    
    anomaly_counts = df_filtered['Anomaly_Score'].value_counts().rename({1: 'Normal', -1: 'Anomaly'})
    anomaly_chart = px.bar(anomaly_counts, labels={'value':'Count', 'index':'Anomaly Score'})
    st.plotly_chart(anomaly_chart)

    st.markdown("""
    **🔎 Insights:**
    - Unsupervised anomaly detection finds points that deviate from expected patterns.
    - Isolation Forest flags potential fraud candidates not seen during supervised training.
    - Useful tool for finding novel fraud cases or internal fraud.
    """)

# ========================
# TAB 6 — 3D PCA Scatterplot
# ========================
with tab6:
    st.header("🧪 Interactive 3D PCA Scatterplot")

    fraud_df = df_filtered[df_filtered['Class'] == 1]
    legit_sample = df_filtered[df_filtered['Class'] == 0].sample(n=2000, random_state=42)
    fraud_oversample = pd.concat([fraud_df] * 5, ignore_index=True)
    balanced_df = pd.concat([legit_sample, fraud_oversample])
    balanced_df['Class_Label'] = balanced_df['Class'].map({0: 'Legitimate', 1: 'Fraud'})

    fig3d = px.scatter_3d(
        balanced_df, x='V1', y='V2', z='V3',
        color='Class_Label',
        symbol='Class_Label',
        size=balanced_df['Class'].map({0: 3, 1: 6}),
        size_max=10, opacity=0.8,
        hover_data=['Amount', 'Time'],
        title='3D PCA Scatterplot'
    )
    fig3d.update_layout(width=900, height=700)
    st.plotly_chart(fig3d)

    st.markdown("""
    **🔎 Insights:**
    - Fraud often clusters differently than legitimate transactions in PCA-reduced space.
    - 3D projection provides intuitive spatial separation for early feature exploration.
    """)

# ========================
# TAB 7 — Dataset Explorer
# ========================
with tab7:
    st.header("📂 Filtered Dataset Explorer")
    st.dataframe(df_filtered)

    csv = df_filtered.to_csv(index=False).encode('utf-8')
    st.download_button(
        label="💾 Download Current Filtered Dataset",
        data=csv,
        file_name='filtered_fraud_data.csv',
        mime='text/csv',
    )
````

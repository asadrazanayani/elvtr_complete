# Step 7 ‚Äî Outlier Detection and Capping (IQR Method)

Highly skewed features with extreme outliers can distort certain machine learning models and reduce stability.

**Why this matters:**
- Outliers may dominate distance-based models or tree splits.
- Capping extreme values helps models focus on the central predictive range.
- We want to cap true outliers but preserve legitimate high-value transactions.

In this step, we will apply **IQR-based outlier capping** to the `amount` feature.

**Capping Rule:**

- Lower Bound = Q1 - 1.5 * IQR
- Upper Bound = Q3 + 1.5 * IQR
- Values beyond these bounds will be capped to the thresholds.

This allows us to reduce the impact of extreme values while preserving overall data integrity.


```python
# Calculate IQR boundaries for amount
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: {Q1}, Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Apply capping
df['amount_capped'] = df['amount'].clip(lower=lower_bound, upper=upper_bound)

# Compare original vs capped distributions
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df['amount'], bins=50, color='skyblue', label='Original', alpha=0.6)
sns.histplot(df['amount_capped'], bins=50, color='red', label='Capped', alpha=0.4)
plt.title('Transaction Amount Distribution: Original vs Capped')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Step 11 ‚Äî Scaling and Train/Test Split

Before training supervised machine learning models, we need to:

3Ô∏è‚É£ **Apply feature scaling**

- Tree-based models (e.g. Random Forest) are generally scale-insensitive, but scaling is still best practice if we plan to experiment with multiple model types.

**Important:**  
- Categorical features encoded as 0/1 (from one-hot encoding) do not require scaling.
- Only continuous features will be scaled.

---

**Features to Scale:**

- `amount`
- `amount_capped`
- `hour_of_day`
- `day_of_week`

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# 1Ô∏è‚É£ Separate features and target
X = df_encoded.drop('is_fraud', axis=1)
y = df_encoded['is_fraud']

# 2Ô∏è‚É£ Train/Test Split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
print("Fraud rate in training set:", round(y_train.mean()*100, 4), "%")
print("Fraud rate in testing set:", round(y_test.mean()*100, 4), "%")

# 3Ô∏è‚É£ Scaling continuous features
continuous_features = ['amount', 'amount_capped', 'hour_of_day', 'day_of_week',
                       'device_id_freq', 'ip_address_freq']

scaler = StandardScaler()

# Fit on training data only
X_train[continuous_features] = scaler.fit_transform(X_train[continuous_features])

# Apply same transformation to test data
X_test[continuous_features] = scaler.transform(X_test[continuous_features])

# Quick verification
print("\nPreview of scaled training data:")
display(X_train.head())
```

# Step 12 ‚Äî Addressing Class Imbalance with SMOTE

Given the extreme class imbalance (~1.19% fraud rate), supervised models may struggle to learn meaningful fraud patterns without resampling.

**Why this matters:**

- Without balancing, models often predict only the majority class (non-fraud) and achieve deceptively high accuracy while failing to detect fraud.
- Oversampling techniques like SMOTE help by synthetically generating new fraud cases that reflect the feature space structure of existing minority class observations.

---

### SMOTE (Synthetic Minority Oversampling Technique):

- Generates synthetic minority samples by interpolating between existing fraud cases.
- Helps models better define the fraud decision boundary without simply duplicating existing records.
- SMOTE should only be applied to the **training set** to avoid data leakage.

---

**Important:**  
- We will apply SMOTE after scaling and splitting to avoid information bleeding.
- We leave the test set fully untouched for unbiased evaluation.


```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE only on training data
smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Verify new class distribution after SMOTE
import collections

counter_before = collections.Counter(y_train)
counter_after = collections.Counter(y_train_smote)

print("Class distribution BEFORE SMOTE:", counter_before)
print("Class distribution AFTER SMOTE:", counter_after)
```

| Class | Before SMOTE | After SMOTE |
|-------|---------------|-------------|
| Non-Fraud (0) | 20,795 | 20,795 |
| Fraud (1)     | 251    | 20,795 |

- SMOTE has successfully balanced the training dataset to a **50/50 class distribution**.
- The minority fraud class has been synthetically oversampled by generating new synthetic records derived from the original 251 fraud cases.

---

### Professional Insight:

- **Why we balance only the training set:**
  - Prevents data leakage into model evaluation.
  - Ensures that model performance metrics remain realistic on the original, imbalanced test set.

- **SMOTE preserves structure:**
  - SMOTE interpolates between existing fraud cases, creating new synthetic samples that are similar but not exact duplicates.
  - This helps the model learn a better decision boundary while avoiding overfitting to repeated observations.

- **Real-world production note:**
  - While SMOTE is excellent for model training and research, production fraud systems often supplement oversampling with domain-informed rule triggers, network-based behavioral profiling, or cost-sensitive algorithms that directly penalize false negatives more heavily.

# Step 16 ‚Äî Controlled Undersampling (90/10 Class Balance)

In the previous step, full undersampling to a 50/50 class balance significantly increased fraud recall but introduced extreme instability, severely degrading precision and generalization.

In real-world fraud detection systems, it is often preferable to apply **controlled partial undersampling** to improve minority class learning while retaining as much majority class information as possible.

---

### Why Controlled Undersampling Matters:

- Preserves majority class diversity and normal behavior variability.
- Allows the model to learn minority class patterns without fully discarding majority data.
- Produces a more stable, generalizable fraud model with better operational potential.

---

### Approach for This Step:

- Undersample the majority class to achieve a **90/10 class balance** in the training set.
- This maintains sufficient normal behavior representation while encouraging stronger fraud detection learning.
- The test set remains fully untouched to ensure unbiased evaluation.

This method reflects standard practices in production fraud modeling pipelines.



```python
# Recombine training features and target for controlled undersampling
train_data = pd.concat([X_train, y_train], axis=1)

# Separate majority and minority classes
df_majority = train_data[train_data['is_fraud'] == 0]
df_minority = train_data[train_data['is_fraud'] == 1]

# Determine undersample size to achieve 90/10 balance
desired_majority_count = int(len(df_minority) * 9)  # 90% majority, 10% minority

# Undersample majority class
df_majority_under = df_majority.sample(n=desired_majority_count, random_state=42)

# Combine undersampled majority with full minority class
train_undersampled_90_10 = pd.concat([df_majority_under, df_minority], axis=0)

# Shuffle dataset
train_undersampled_90_10 = train_undersampled_90_10.sample(frac=1, random_state=42)

# Separate back into features and target
X_train_under_90_10 = train_undersampled_90_10.drop('is_fraud', axis=1)
y_train_under_90_10 = train_undersampled_90_10['is_fraud']

# Verify class distribution
print("Class distribution after 90/10 undersampling:")
print(y_train_under_90_10.value_counts())

# Retrain Random Forest on controlled undersampled data
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, RocCurveDisplay

rf_under_90_10 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_under_90_10.fit(X_train_under_90_10, y_train_under_90_10)

# Predict on untouched test set
y_pred_under_90_10 = rf_under_90_10.predict(X_test)
y_proba_under_90_10 = rf_under_90_10.predict_proba(X_test)[:, 1]

# Evaluate performance
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_under_90_10))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_under_90_10, digits=4))

auc_under_90_10 = roc_auc_score(y_test, y_proba_under_90_10)
print(f"\nROC AUC Score: {auc_under_90_10:.4f}")

# ROC Curve
import matplotlib.pyplot as plt
RocCurveDisplay.from_estimator(rf_under_90_10, X_test, y_test)
plt.title("Random Forest with 90/10 Undersampling - ROC Curve")
plt.show()
```

# Step 17 ‚Äî Threshold Selection using Lift Table and Cumulative Gains Curve

In fraud detection, selecting an appropriate probability threshold is one of the most important modeling decisions.

By default, classifiers assign fraud labels to predictions where the model probability exceeds 0.5.  
However, this threshold is often inappropriate for rare-event problems like fraud detection.

---

### Why Threshold Tuning is Critical:

- Fraud models typically output probability scores well below 0.5 for most observations.
- Applying a fixed 0.5 threshold often produces excellent overall accuracy but fails to capture meaningful fraud cases.
- **Business-aligned thresholds** are usually selected by analyzing:
  - Cumulative Gains / Lift Charts
  - Recall vs. review volume trade-offs
  - False positive management capacity

---

### What a Lift Table Shows:

- Sorts all test set predictions by predicted fraud probability (highest to lowest).
- Divides the data into deciles (or percentiles).
- Measures the cumulative capture of fraud cases as more high-risk transactions are flagged.
- Helps determine what percentage of total transactions need to be reviewed to capture a desired proportion of fraud cases.

This approach is used extensively in operational fraud analytics to align models with real-world resource allocation and business risk appetite.


```python
import pandas as pd
import numpy as np

# Build dataframe with test set predictions
df_lift = pd.DataFrame({
    'y_true': y_test,
    'y_proba': y_proba_under_90_10  # Use latest model scores here
})

# Sort by predicted probability descending
df_lift = df_lift.sort_values(by='y_proba', ascending=False).reset_index(drop=True)

# Assign decile ranking
df_lift['decile'] = pd.qcut(df_lift.index, 10, labels=False)

# Calculate cumulative fraud capture per decile
lift_table = df_lift.groupby('decile').agg({
    'y_true': ['sum', 'count']
}).reset_index()

lift_table.columns = ['decile', 'frauds_captured', 'total_cases']

# Calculate cumulative sums
lift_table['cumulative_fraud'] = lift_table['frauds_captured'].cumsum()
lift_table['cumulative_cases'] = lift_table['total_cases'].cumsum()

# Calculate cumulative fraud capture rate (recall)
total_frauds = df_lift['y_true'].sum()
lift_table['cumulative_recall'] = lift_table['cumulative_fraud'] / total_frauds

# Calculate review burden (total reviewed vs. total population)
total_cases = len(df_lift)
lift_table['review_rate'] = lift_table['cumulative_cases'] / total_cases

# Display lift table
print("\nLift Table:")
display(lift_table[['decile', 'cumulative_fraud', 'cumulative_recall', 'review_rate']])

# Plot cumulative gains curve
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(lift_table['review_rate'], lift_table['cumulative_recall'], marker='o')
plt.plot([0,1], [0,1], linestyle='--', color='gray', label='Random')
plt.title("Cumulative Gains Curve")
plt.xlabel("Percentage of Population Reviewed")
plt.ylabel("Cumulative Recall (Fraud Captured)")
plt.legend()
plt.grid(True)
plt.show()
```

### Cumulative Gains Curve Summary:

- The model captures approximately:
  - 26% of fraud cases by reviewing the top 10% of predictions.
  - 49% of fraud cases by reviewing the top 20%.
  - 73% of fraud cases by reviewing the top 30%.
  - 99% of fraud cases by reviewing the top 40%.

- After the top 40% of predictions, nearly all fraud cases have been identified, and the curve flattens.

---

### Interpretation:

- This model exhibits reasonably strong **ranking power** ‚Äî most fraud cases are concentrated in the higher-scoring segments of the population.
- While precision remains low at any given cutoff, **rank-ordering is effective**, which is often the primary goal in fraud scoring systems.
- This enables business teams to set appropriate operational thresholds based on fraud risk appetite and available review capacity.

---

### Threshold Selection Considerations:

  - Often, thresholds are set to capture **60‚Äì80% of fraud** while limiting review workload to manageable levels.

# Step 18 ‚Äî Deriving the Optimal Probability Threshold

Now that we have analyzed the lift table and cumulative gains curve, we can convert the selected review volume into a concrete model probability cutoff.

---

### Approach:

- Sort all test set predictions by predicted probability (descending).
- Select the top X% of cases corresponding to the target review rate (e.g., top 30% for ~73% fraud capture).
- Extract the minimum probability score within that top segment.
- This becomes the new threshold for operational deployment.

---

### Business Example:

- Suppose the fraud operations team is willing to review approximately **30% of the total population**.
- From the lift table, this corresponds to ~73% fraud capture.
- We will now determine the actual model score cutoff that defines this threshold.


```python
# Desired review rate target (e.g., top 30% of records reviewed)
target_review_rate = 0.30

# Determine number of records to select based on target review volume
n_review = int(len(df_lift) * target_review_rate)

# Get the minimum score in top X% of records
threshold = df_lift.loc[:n_review-1, 'y_proba'].min()

print(f"Suggested probability threshold for top {int(target_review_rate*100)}% review: {threshold:.4f}")

# Apply new threshold to calculate adjusted confusion matrix
y_pred_adjusted = (y_proba_under_90_10 >= threshold).astype(int)

from sklearn.metrics import confusion_matrix, classification_report

print("\nConfusion Matrix with adjusted threshold:")
print(confusion_matrix(y_test, y_pred_adjusted))

print("\nClassification Report with adjusted threshold:")
print(classification_report(y_test, y_pred_adjusted, digits=4))
```
